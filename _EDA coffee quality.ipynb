```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pandas_profiling import ProfileReport
import seaborn as sns
import csv
import numpy as np
%matplotlib inline
```

```python
coffee_quality_analysis= pd.read_csv('coffee_data_cleaned.csv')
```

```python
coffee_quality_analysis = coffee_quality_analysis.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis = 1)             
```

```python
coffee_quality_analysis.head()
```

```python
#multivariate variate analysis

coffee_quality_analysis = coffee_quality_analysis.profile_report()
```

```python
coffee_quality_analysis.to_file("coffee_quality_report.html")
```

```python
#univariate analysis
#variables to analyze: harvest year, country of origin, processing method
#variables to analyze with: aroma, flavor,aftertaste, acidity, body and balance
coffee_quality_analysis2= pd.read_csv('coffee_data_cleaned.csv')
```

```python
coffee_quality_analysis3 = coffee_quality_analysis2.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis = 1)             
```

```python
sns.scatterplot(y='Flavor', x='Total.Cup.Points', data=coffee_quality_analysis3)
plt.xticks(rotation = 45)
```

```python
plt.figure(figsize = (20,8))
#annot= annotation
sns.heatmap(coffee_quality_analysis3.corr(), annot = True)
```
